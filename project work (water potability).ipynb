# Checking whether the columns ph, Hardness, Solids, Chloramines, Sulfate, Conductivity, Organic_carbon, Trihalomethanes and Turbidity are related with column Potability or not.


* Since all the columns are continous and they can contain outliers and none of them are discrete, so, let's check the relation by using median and boxplot.


```python
df.head()
```

# Ph

```python

sns.boxplot(data = df, x = 'Potability', y = 'ph')
plt.show()
```

```python
df.groupby('Potability').ph.median()
```

* Column ph contains null values.So, now fill those null values using Potability.


```python
def fill_missing_ph(cols):

    ph = cols[0]
    Potability = cols[1]

    if pd.isna(ph):

        if Potability == 1:
            return 7.036752

        else:
            return 7.035456

    return ph

```

```python
df.ph = df[['ph', 'Potability']].apply(fill_missing_ph, axis = 1)
```

```python
df.isna().sum()
```

* Null values has been filled except Sulfate and Trihalomethanes

# Since sulfate column also contains null value. So fill the null values using potability.


```python
def fill_missing_sulfate(cols):

    Sulfate = cols[0]
    Potability = cols[1]

    if pd.isna(Sulfate):

        if Potability == 1:
            return 331.838167

        else:
            return 333.389426

    return Sulfate
```

```python
df.Sulfate = df[['Sulfate', 'Potability']].apply(fill_missing_sulfate, axis = 1)
```

# Trihalomethanes


```python
sns.boxplot(data=df , x='Potability', y='Trihalomethanes')
plt.show()
```

```python
df.groupby('Potability').Trihalomethanes.median()
```

* Since Trihalomethanes column contains null value.So we must fill missing values


```python
def fill_missing_trihalomethanes(cols):

    Trihalomethanes = cols[0]
    Potability = cols[1]

    if pd.isna(Trihalomethanes):

        if Potability == 1:
            return 66.678214

        else:
            return 66.542198

    return Trihalomethanes
```

```python
df.Trihalomethanes = df[['Trihalomethanes', 'Potability']].apply(fill_missing_trihalomethanes, axis = 1)
```

# Traine test split

```python
from sklearn.model_selection import train_test_split
```

```python
X = df.iloc[ : , : -1]
Y = df.iloc[ : , -1]
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size = 0.2, random_state = 42)
```

```python
X_train.shape, Y_train.shape
```

```python
X_test.shape, Y_test.shape
```

# Scaling the feature/X usind StandardScaler


```python
from sklearn.preprocessing import StandardScaler
```

```python
sc_train = StandardScaler()
X_train_scaled = sc_train.fit_transform(X_train)

sc_test = StandardScaler()
X_test_scaled = sc_train.fit_transform(X_test)
```

# We will use 2 classification algorithms

* Support Vector Machine

* Logistic Regression


#SVM model

```python
sns.pairplot(data = df, hue = 'Potability')
plt.show()
```

```python
from sklearn.svm import SVC
```

```python
SVM_model = SVC()
SVM_model.fit(X_train_scaled, Y_train)
Y_pred_SVM = SVM_model.predict(X_test_scaled)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
accuracy_SVM = accuracy_score(Y_test, Y_pred_SVM)
```

```python
recall_SVM = recall_score(Y_test, Y_pred_SVM)
```

```python
precision_SVM = precision_score(Y_test, Y_pred_SVM)
```

```python
f1_SVM = f1_score(Y_test, Y_pred_SVM)
```

```python
print(f'''The accuracy score of the SVM model is: {accuracy_SVM} i.e. {accuracy_SVM * 100} %\n
The recall score of the SVM model is: {recall_SVM} i.e. {recall_SVM * 100} %\n
The precision score of the SVM model is: {precision_SVM} i.e. {precision_SVM * 100} %\n
The f1 score of the SVM model is: {f1_SVM} i.e. {f1_SVM * 100} %''')
```

```python
cm_SVM = confusion_matrix(Y_test, Y_pred_SVM)
```

```python
fig_SVM = ConfusionMatrixDisplay(cm_SVM, display_labels = SVM_model.classes_)
fig_SVM.plot()
plt.show()
```

# Logistic Regression model

```python
from sklearn.linear_model import LogisticRegression
```

```python
LR_model = LogisticRegression(max_iter = 5000)
LR_model.fit(X_train_scaled, Y_train)

Y_pred_LR = LR_model.predict(X_test_scaled)
```

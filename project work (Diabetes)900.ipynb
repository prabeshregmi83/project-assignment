# Train Test split

```python
# Importing libraries
from sklearn.model_selection import train_test_split
```

```python
df.head(2) # Using 80% data for training and 20% for test
```

```python
X = df.iloc[ : , : -1]
Y = df.iloc[ : , -1]
```

```python
X_train , X_test , Y_train ,Y_test = train_test_split(X , Y , test_size = 0.2 , random_state = 42)
```

```python
X_train.shape , X_test.shape
```

```python
Y_train.shape ,Y_test.shape
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression(max_iter = 5000)
model.fit(X_train, Y_train)

Y_pred = model.predict(X_test)
```

# Evaluation

```python
from sklearn.metrics import accuracy_score, recall_score, precision_score, f1_score, confusion_matrix, ConfusionMatrixDisplay
```

```python
accuracy = accuracy_score(Y_test, Y_pred)
```

```python
recall = recall_score(Y_test, Y_pred)
```

```python
precision = precision_score(Y_test , Y_pred)
```

```python
f1 = f1_score(Y_test , Y_pred) # f1_score is ML evaluation metric that measures model's accuracy
```

```python

print(f'''The accuracy score of the model is: {accuracy} i.e. {accuracy * 100} %\n
The recall score of the model is: {recall} i.e. {recall * 100} %\n
The precision score of the model is: {precision} i.e. {precision * 100} %\n
The f1 score of the model is: {f1} i.e. {f1 * 100} %''')

```

```python
cm = confusion_matrix(Y_test , Y_pred)
```

```python
fig = ConfusionMatrixDisplay(cm , display_labels = model.classes_)
fig.plot()
plt.show()
```
